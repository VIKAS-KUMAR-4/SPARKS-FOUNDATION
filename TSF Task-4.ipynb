Firstly, we will import several dependencies such as pandas, numpy, and seaborn.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

df = pd.read_csv('Iris.csv')
```

**Now, Splitting the Data set**

```python
all_inputs = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
all_classes = df['Species'].values

(train_inputs, test_inputs, train_classes, test_classes) = train_test_split(all_inputs, all_classes, train_size=0.7, random_state=1)
```

**Trainiing the dataset**

```python
decision_tree=DecisionTreeClassifier()
decision_tree.fit(train_inputs, train_classes)
```

**Making Prediction On Test data**

```python
y_pred=decision_tree.predict(test_inputs)
y_pred
```
